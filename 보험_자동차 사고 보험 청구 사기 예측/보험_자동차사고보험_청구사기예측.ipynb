````python
# Î™®Îç∏ÎßÅÏùÑ ÌïôÏäµÌïòÍ∏∞ ÏúÑÌïú Feature(X)ÏôÄ YÎç∞Ïù¥ÌÑ∞Î•º Íµ¨Î∂ÑÌïòÎäî Îã®Í≥Ñ
# monthÎäî ÏÉàÎ°úÏö¥ ÌååÏÉù Î≥ÄÏàòÎ°úÏÑú Ï∂îÍ∞Ä

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

X=df.drop(['claim_number','claim_date','year','age_of_driver_gp','annual_income_gp','fraud'],axis=1)
Y=df['fraud']

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,stratify=Y)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
````

````python
categorical_list
````

````python
# ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÏùÑ Categorical Î≥ÄÏàòÎäî Î¶¨Ïä§Ìä∏ÏóêÏÑú Ï†úÏô∏
categorical_list.remove('claim_date')
````

````python
# ÏÉàÎ°ú Ï∂îÍ∞ÄÌïú Î≥ÄÏàòÎäî ÏóÜÍ∏∞ ÎïåÎ¨∏Ïóê Ï∂îÍ∞Ä
categorical_list.append('month')
categorical_list
````

````python
df[categorical_list]
````

````python
#Categorical(Î≤îÏ£ºÌòï) Î≥ÄÏàòÎäî One-hot-encoding or Label-encodingÏùÑ ÌÜµÌï¥ Ïà´ÏûêÌòï Î≥ÄÏàòÎ°ú Î≥ÄÍ≤ΩÌï¥ÏïºÌï®
#One-hot-encodingÏùÄ Ï∞®ÏõêÏù¥ ÎßåÌù• Î≥ÄÏàòÏóêÎäî Î∂àÎ¶¨, Label-encodingÏùÄ ÌöåÍ∑ÄÍ¥ÄÎ†® ÏïåÍ≥†Î¶¨Ï¶òÏóêÏÑúÎäî ÏÇ¨Ïö© Ïñ¥Î†§ÏõÄ (Tree Í≥ÑÏó¥ ÏïåÍ≥†Î¶¨Ï¶òÏóêÏÑúÎäî ÏÇ¨Ïö© Í∞ÄÎä•)
from sklearn.preprocessing import LabelEncoder

for col in categorical_list:
    print(col)
    le=LabelEncoder()
    le.fit(list(x_train[col].values) + list(x_test[col].values))
    x_train[col] = le.transform(x_train[col])
    x_test[col] = le.transform(x_test[col])
````

````python
x_train[categorical_list].head(3)
````

````python
x_test[categorical_list].head(3)
````

### ‚àô Î™®Îç∏ ÌïôÏäµ Î∞è ÌèâÍ∞Ä
---
* Î™®Îç∏ÏùÑ ÌïôÏäµÌïòÍ≥† ÏòàÏ∏°ÏùÑ ÏàòÌñâÌïòÏó¨ ÏÑ±Îä•ÏùÑ ÌèâÍ∞Ä 

````python
import lightgbm as lgb

# Hyper parametre setting

d_train= lgb.Dataset(x_train, label=y_train)

params={}
params ['learning_rate'] =0.02
params['boosting_type'] = 'gbdt' # GradientBoostingDecisionTree
params['objective'] = 'binary'
params['metric'] = 'binary_logloss' #metric for binary-class
params['max_depth'] =2
params['num_leaves'] = 4 # ÏµúÎåÄ leavesÎäî 2^(max_depth)
params['seed']=23456

# ÌïôÏäµ
clf=lgb.train(params,d_train,500) #1000 epocsÏóêÏÑú Î™®Îç∏ ÌõàÎ†®
````

````python
from sklearn.metrics import classification_report

y_pred_train=clf.predict(x_train)
for i in range(0,len(y_pred_train)):
    if y_pred_train[i]>=.3:
        y_pred_train[i]=1
    else:
        y_pred_train[i]=0
y_pred_test = clf.predict(x_test)
for i in range(0,len(y_pred_test)):
    if y_pred_test[i]>=.3:
        y_pred_test[i]=1
    else:
        y_pred_test[i]=0
print(classification_report(y_train,y_pred_train))
print(classification_report(y_test,y_pred_test))
````

````python
from sklearn.metrics import roc_curve
def roc_curve_plot(y_test,pred_proba_c1):
    #ÏûÑÍ≥óÍ∞íÏóê Îî∞Î•∏ FPR,TPR Í∞íÏùÑ Î∞òÌôò Î∞õÏùå.
    #FPR : ÏïîÌôòÏûêÍ∞Ä ÏïÑÎãå ÌôòÏûêÎ•º ÏïîÌôòÏûêÎùºÍ≥† Ïûò Î™ª ÏòàÏ∏°Ìïú ÎπÑÏú®
    #TPR : Recall
    fprs, tprs , threshols= roc_curve(y_test,pred_proba_c1)

    # ROC CurveÎ•º plt Í≥°ÏÑ†ÏúºÎ°ú Í∑∏Î¶º.
    plt.plot(fprs, tprs, label='ROC')
    #Í∞ÄÏö¥Îç∞ ÎåÄÍ∞ÅÏÑ† ÏßÅÏÑ†ÏùÑ Í∑∏Î¶º
    plt.plot([0,1],[0,1],'k--' , label='Random', color='red')

    #FPR X Ï∂ïÏùò ScaleÏùÑ 0.1 Îã®ÏúÑÎ°ú Î≥ÄÍ≤Ω X,Y Ï∂ïÎ™Ö ÏÑ§Ï†ïÎì±
    start, end= plt.xlim()
    plt.xticks(np.round(np.arange(start,end,0.1),2))
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xlabel('FPR(1 - Sensitivity)')
    plt.ylabel('TPR(Recall)')
    plt.legend()
    plt.show()
````
